```python
#Importing the necessary libraries
import numpy as np
import matplotlib.pyplot as plt

#Identifying the X values which are the irradiance related temperature difference values for the 9 measurement
x = np.array([0.051658322, 0.086182873, 0.08052826 ,0.060324618
             ,0.101224791,0.09920733 ,0.058737417 ,0.038981923 ,0.023973205])

#Identifying the Y values which are the efficiency values for the 9 measurement
y= np.array([35.3476789 ,23.8521678 ,28.074719 ,29.6874419
            ,19.3887957,42.4757266,25.9402341 ,35.7805957 ,36.8581919])
 
#Preparing the figure setup    
fig,ax = plt.subplots()

#Scatter plot for the experiment measurements
ax.scatter(x, y)

#Formatting the graph
ax.set_ylabel('Efficiency η [%] ',fontweight="bold" )
ax.set_xlabel('(Tm -Ta)/G  [(m^2 . k) / W]',fontweight="bold" )
ax.set(xlim= (0,0.12), ylim= (0,80))
ax.tick_params(direction="in")
ax.set_title('Solar collector characteristics η = f((Tm − Ta)/G)', fontname="Times New Roman", size=14,fontweight="bold")

#line fitting
z = np.polyfit(x,y,1)
p = np.poly1d(z)
ax.plot(x,p(x), color = "g")

plt.show()
```
